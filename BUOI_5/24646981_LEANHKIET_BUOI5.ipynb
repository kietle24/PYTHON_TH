```python
import pandas as pd 
df = pd.read_csv("population_data_standard.csv")
```

```python
df["TangTruong"] = df.groupby("QuocGia")["DanSo"].pct_change()*100 
print(df)
```

```python
#Tìm ra những quốc gia có tốc độ tăng trưởng cao nhất và thấp nhất.
quocgia_tangtruong = df.groupby("QuocGia")["TangTruong"].mean() 
max_country = quocgia_tangtruong.idxmax()
min_country = quocgia_tangtruong.idxmin() 
print("Quốc gia có tốc độ tăng trưởng TB cao nhất:", max_country, quocgia_tangtruong[max_country])
print("Quốc gia có tốc độ tăng trưởng TB thấp nhất:", min_country, quocgia_tangtruong[min_country])
```

```python
# Vẽ biểu đồ thể hiện sự thay đổi dân số theo thời gian của 5 quốc gia có dân số lớn nhất
import matplotlib.pyplot as plt

# Lấy 5 quốc gia có dân số lớn nhất năm 2024
top5_2024 = df[df["Nam"] == 2024].sort_values(by="DanSo", ascending=False).head(5)
top5_country = top5_2024["QuocGia"].tolist()

# Lọc dữ liệu của 5 quốc gia này qua các năm
df_top5 = df[df["QuocGia"].isin(top5_country)]

plt.figure(figsize=(12,6))
for country in top5_country:
    subnet = df_top5[df_top5["QuocGia"] == country]
    plt.plot(subnet["Nam"], subnet["DanSo"], marker='o', label=country)
plt.title("Sự Thay Đổi Dân Số Của 5 Quốc Gia Lớn Nhất")
plt.xlabel("Năm")
plt.ylabel("Dân Số")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Chuyển cột 'Ngay' về kiểu datetime và đặt làm index
import pandas as pd 
df["Ngay"] = pd.to_datetime(df["Ngay"]) 
df = df.set_index("Ngay")
# Tính theo tuần
weekly = df.resample("W").agg({"DongCua": ["mean", "max", "min"]})
#tinh theo thang 
monthly = df.resample("M").agg({"DongCua": ["mean", "max", "min"]})
print(weekly)
print(monthly)
```

```python
# % thay đổi giá
df["Return"] = df["DongCua"].pct_change()

# log return
import numpy as np
df["LogReturn"] = np.log(df["DongCua"] / df["DongCua"].shift(1))
threshold = 2 * df["Return"].std()
df["BigMove"] = df["Return"].abs() > threshold
big_moves = df[df["BigMove"]]
print("Những ngày biến động lớn:")
print(big_moves[["DongCua", "Return"]])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Giả lập dữ liệu 100 ngày (ngày giao dịch)
dates = pd.date_range(start="2024-01-01", periods=100, freq="B")  # Business days
prices = np.cumsum(np.random.normal(0, 1, size=len(dates))) + 100

df = pd.DataFrame({"Date": dates, "Close": prices})
df.set_index("Date", inplace=True)
df["MA20"] = df["Close"].rolling(window=20).mean()
df["MA50"] = df["Close"].rolling(window=50).mean()
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Close"], label="Giá đóng cửa", linewidth=2)
plt.plot(df.index, df["MA20"], label="MA20 (ngắn hạn)", linestyle="--")
plt.plot(df.index, df["MA50"], label="MA50 (dài hạn)", linestyle="--")

plt.title("Giá cổ phiếu & Đường trung bình động (MA20, MA50)")
plt.xlabel("Ngày")
plt.ylabel("Giá")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import csv

FILE_NAME = "library.csv"

# 1. Nhập và lưu thông tin sách
def add_book():
    ten = input("Nhập tên sách: ")
    tac_gia = input("Nhập tác giả: ")
    nam_xb = input("Nhập năm xuất bản: ")
    the_loai = input("Nhập thể loại: ")

    with open(FILE_NAME, mode="a", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow([ten, tac_gia, nam_xb, the_loai])
    print("Đã lưu thông tin sách vào file.")

# 2. Đọc và hiển thị danh sách sách
def show_books():
    try:
        with open(FILE_NAME, mode="r", encoding="utf-8") as file:
            reader = csv.reader(file)
            print(f"{'Tên sách':30} | {'Tác giả':20} | {'Năm XB':6} | {'Thể loại':15}")
            print("-" * 80)
            for row in reader:
                if row:  # tránh dòng rỗng
                    print(f"{row[0]:30} | {row[1]:20} | {row[2]:6} | {row[3]:15}")
    except FileNotFoundError:
        print("Chưa có dữ liệu. Hãy thêm sách trước.")

# 3. Tìm kiếm sách
def search_books():
    keyword = input("Nhập từ khóa tìm kiếm: ").lower()
    try:
        with open(FILE_NAME, mode="r", encoding="utf-8") as file:
            reader = csv.reader(file)
            found = False
            for row in reader:
                if any(keyword in field.lower() for field in row):
                    print(f"Tìm thấy: {row}")
                    found = True
            if not found:
                print("Không tìm thấy sách phù hợp.")
    except FileNotFoundError:
        print("Chưa có dữ liệu.")

# Menu chính
def main():
    while True:
        print("\n=== QUẢN LÝ THƯ VIỆN ===")
        print("1. Thêm sách")
        print("2. Hiển thị danh sách sách")
        print("3. Tìm kiếm sách")
        print("4. Thoát")
        choice = input("Chọn chức năng: ")

        if choice == "1":
            add_book()
        elif choice == "2":
            show_books()
        elif choice == "3":
            search_books()
        elif choice == "4":
            break
        else:
            print("Lựa chọn không hợp lệ.")

if __name__ == "__main__":
    main()
```

```python
Bài tập 8: Phân tích dữ liệu năng lượng tiêu thụ
Mô tả: Phân tích mức tiêu thụ điện của một gia đình trong một năm.
1.	Đọc dữ liệu tiêu thụ điện từ file CSV (dữ liệu theo tháng).
2.	Tính toán tổng lượng điện tiêu thụ trong năm, lượng điện tiêu thụ trung bình mỗi tháng.
3.	Tìm ra tháng có mức tiêu thụ điện cao nhất, thấp nhất.
4.	Vẽ biểu đồ thể hiện mức tiêu thụ điện trong từng tháng và trực quan hóa xu hướng tiêu thụ điện theo thời gian.
5.	Vẽ lưu đồ giải thuật
Gợi ý:
•	Lưu đồ mô tả quy trình: Đọc dữ liệu năng lượng, tính toán mức tiêu thụ, tìm tháng tiêu thụ cao nhất, vẽ biểu đồ.
•	Yêu cầu: Vẽ lưu đồ giải thuật thể hiện quy trình phân tích dữ liệu tiêu thụ điện và trực quan hóa sự thay đổi theo tháng.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Thay đổi đường dẫn nếu cần
FILE_PATH = "energy.csv"

def load_data(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Không tìm thấy file: {file_path}")
    # Đọc CSV; cố gắng parse cột Month thành datetime
    df = pd.read_csv(file_path)
    if "Month" not in df.columns or "Consumption" not in df.columns:
        raise ValueError("File CSV phải có cột 'Month' và 'Consumption'")
    # Chuyển Month sang datetime (nếu có định dạng YYYY-MM hoặc YYYY-MM-DD)
    try:
        df["Month"] = pd.to_datetime(df["Month"], format="%Y-%m")
    except Exception:
        df["Month"] = pd.to_datetime(df["Month"], errors="coerce")
    # Chuyển Consumption sang numeric, thay giá trị không hợp lệ bằng NaN
    df["Consumption"] = pd.to_numeric(df["Consumption"], errors="coerce")
    # Sắp xếp theo thời gian
    df = df.sort_values("Month").reset_index(drop=True)
    return df

def compute_statistics(df):
    # Bỏ các dòng có Consumption là NaN trước khi tính
    valid = df.dropna(subset=["Consumption"])
    total = valid["Consumption"].sum()
    monthly_avg = valid["Consumption"].mean()
    # Tháng max/min (lấy lần xuất hiện đầu nếu trùng)
    idx_max = valid["Consumption"].idxmax()
    idx_min = valid["Consumption"].idxmin()
    month_max = valid.loc[idx_max, "Month"]
    month_min = valid.loc[idx_min, "Month"]
    val_max = valid.loc[idx_max, "Consumption"]
    val_min = valid.loc[idx_min, "Consumption"]
    return {
        "total_kwh": total,
        "monthly_avg_kwh": monthly_avg,
        "month_max": month_max,
        "val_max": val_max,
        "month_min": month_min,
        "val_min": val_min,
        "n_missing": df["Consumption"].isna().sum()
    }

def plot_consumption(df, rolling_window=3):
    # Chuẩn bị dữ liệu cho biểu đồ (bao gồm NaN để hiển thị thiếu dữ liệu)
    plot_df = df.set_index("Month")
    plt.figure(figsize=(10,5))
    plt.plot(plot_df.index, plot_df["Consumption"], marker='o', label="Tiêu thụ (kWh)")
    # Đường trung bình động để thấy xu hướng
    rolling = plot_df["Consumption"].rolling(window=rolling_window, min_periods=1).mean()
    plt.plot(plot_df.index, rolling, linestyle='--', marker=None, label=f"{rolling_window}-tháng trung bình động")
    plt.title("Mức tiêu thụ điện theo tháng")
    plt.xlabel("Tháng")
    plt.ylabel("Tiêu thụ (kWh)")
    plt.grid(axis='y', linestyle=':', linewidth=0.5)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

def main():
    try:
        df = load_data(FILE_PATH)
    except Exception as e:
        print("Lỗi khi tải dữ liệu:", e)
        return

    stats = compute_statistics(df)
    # Xuất kết quả thống kê
    print("Thống kê tiêu thụ điện trong dữ liệu:")
    print(f"- Tổng lượng điện trong năm (kWh): {stats['total_kwh']:.2f}")
    print(f"- Lượng điện trung bình mỗi tháng (kWh): {stats['monthly_avg_kwh']:.2f}")
    print(f"- Tháng có mức tiêu thụ cao nhất: {stats['month_max'].strftime('%Y-%m')} với {stats['val_max']} kWh")
    print(f"- Tháng có mức tiêu thụ thấp nhất: {stats['month_min'].strftime('%Y-%m')} với {stats['val_min']} kWh")
    print(f"- Số tháng thiếu/không hợp lệ (Consumption missing): {stats['n_missing']}")

    # Hiển thị bảng dữ liệu (tùy chọn)
    print("\nBảng dữ liệu (hiển thị 15 dòng đầu nếu dài):")
    display_df = df.copy()
    # Format cột Month thành chuỗi YYYY-MM để dễ đọc
    display_df["Month"] = display_df["Month"].dt.strftime("%Y-%m")
    print(display_df.head(15).to_string(index=False))

    # Vẽ biểu đồ
    plot_consumption(df, rolling_window=3)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu từ file CSV
df = pd.read_csv("social_posts.csv", parse_dates=["date"])

# Tạo thêm cột interactions (tổng thích + bình luận + chia sẻ)
df["interactions"] = df["likes"] + df["comments"] + df["shares"]

# 2. Tính tổng số lượt thích, bình luận, chia sẻ trong từng tháng
df["month"] = df["date"].dt.to_period("M")  # lấy tháng dạng YYYY-MM
monthly_totals = df.groupby("month")[["likes", "comments", "shares"]].sum()

print("Tổng lượt thích, bình luận, chia sẻ theo từng tháng:")
print(monthly_totals)

# 3. Phân tích bài viết có số tương tác cao nhất, thấp nhất
max_post = df.loc[df["interactions"].idxmax()]
min_post = df.loc[df["interactions"].idxmin()]

print("\nBài viết có tương tác cao nhất:")
print(max_post)

print("\nBài viết có tương tác thấp nhất:")
print(min_post)

# 4. Vẽ biểu đồ sự thay đổi theo thời gian
plt.figure(figsize=(12,6))
plt.plot(monthly_totals.index.astype(str), monthly_totals["likes"], marker="o", label="Likes")
plt.plot(monthly_totals.index.astype(str), monthly_totals["comments"], marker="o", label="Comments")
plt.plot(monthly_totals.index.astype(str), monthly_totals["shares"], marker="o", label="Shares")

plt.title("Biểu đồ thay đổi lượt tương tác theo tháng")
plt.xlabel("Tháng")
plt.ylabel("Số lượng")
plt.legend()
plt.grid(True)
plt.show()
```
