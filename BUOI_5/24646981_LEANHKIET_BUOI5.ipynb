```python
Bài tập 1: Phân tích dữ liệu điểm thi sinh viên
Mô tả: Bạn có danh sách điểm thi của các sinh viên trong một lớp học.
1.	Đọc dữ liệu từ một file CSV chứa thông tin về mã sinh viên, tên sinh viên, điểm từng môn học.(20 sinh vien cho ngau nhien thong tin(round a file ) )
2.	Tính toán điểm trung bình của từng sinh viên và thêm một cột mới vào dataframe để chứa thông tin này.
3.	Phân tích thống kê về điểm trung bình như giá trị lớn nhất, nhỏ nhất, trung vị, trung bình, phương sai, độ lệch chuẩn.
4.	Vẽ biểu đồ trực quan (histogram, boxplot) để thể hiện phân phối điểm của lớp học.
5.	Vẽ lưu đồ thực hiện:
Gợi ý:
•	Lưu đồ thể hiện quy trình: Đọc dữ liệu, tính toán điểm trung bình, phân tích thống kê, vẽ biểu đồ.
•	Yêu cầu: Vẽ lưu đồ giải thuật để minh họa quy trình phân tích và tính toán điểm thi sinh viên từ dữ liệu CSV.
```

```python
#1.1
import pandas as pd 
df = pd.read_csv("diem_thi.csv")
print(df.head(21))
```

```python
#1.2
# Tính điểm trung bình theo hàng (axis=1)

df["DiemTB"] = df[["Toan","Ly","Hoa"]].mean(1)
print(df.head())

```

```python
#1.3
menu_thong_ke = df["DiemTB"].describe()
diem_tb = df["DiemTB"]
max_diem = diem_tb.max()
min_diem = diem_tb.min()
median_diem = diem_tb.median() #trung vi giua cac gia tri 
mean_diem = diem_tb.mean() 
var_diem = diem_tb.var() # phuong sai 
std_diem = diem_tb.std() # do lech chuan 

print("Điểm cao nhất:", max_diem)
print("Điểm thấp nhất:", min_diem)
print("Trung vị:", median_diem)
print("Trung bình:", mean_diem)
print("Phương sai:", var_diem)
print("Độ lệch chuẩn:", std_diem)


```

```python
#1.4 
#histrogram 
import pandas as pd 
import matplotlib.pyplot as plt 
df = pd.read_csv("diem_thi.csv")
df["DiemTB"] = df[["Toan","Ly","Hoa"]].mean(1)
'''
df["DiemTB"]: dữ liệu cần vẽ, ở đây là điểm trung bình của từng sinh viên
bins=10: chia khoảng giá trị của điểm trung bình thành 10 “thùng” (khoảng tần suất)
color="skyblue": màu thanh của histogram
edgecolor="black": viền thanh màu đen để các thanh phân biệt rõ ràng
'''
plt.hist(df["DiemTB"], bins = 8 , color = "skyblue", edgecolor ="black")
plt.title("bieu do diem trung binh cua sinh vien ")
plt.xlabel("diem trung binh")
plt.ylabel("so sinh vien")
plt.show()
```

```python
#2.3
'''
y tuong : 
Mỗi ngày trong cột Ngay thuộc một tuần nhất định (ISO tuần)
Bạn có thể thêm cột tuần dựa trên ngày:
dt.to_period('W') → chuyển ngày sang khoảng tuần, ví dụ 2025-09-01/2025-09-06
Sau đó, bạn groupby tuần và tính mean của cột SoTien
'''
import pandas as pd 
df = pd.read_csv("chitieu.csv")
df["Ngay"] = pd.to_datetime(df["Ngay"])
df["Tuan"] = df["Ngay"].dt.to_period("W")
tb_tuan = df.groupby("Tuan")["SoTien"].mean().reset_index() 
tb_tuan_columns =["Tuan","TB_ChiTien"]
print(tb_tuan)
```

```python
#3_1 
import pandas as pd 
df = pd.read_csv("weathers.csv")

```

```python
#3_2
for month, group in df.groupby("year_month"):
    thoitiet = group["temp_avg_C"]
    max_tt = thoitiet.max()
    min_tt = thoitiet.min()
    tb_tt = thoitiet.mean()
    print(f"Tháng {month}:")
    print("  Nhiệt độ cao nhất:", max_tt)
    print("  Nhiệt độ thấp nhất:", min_tt)
    print("  Nhiệt độ trung bình:", tb_tt)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Chuyển cột date về kiểu datetime
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
# Tính nhiệt độ trung bình theo tháng
monthly_avg = df.groupby('month')['temp_avg_C'].mean()
# Vẽ biểu đồ cột
plt.figure(figsize=(10,6))
monthly_avg.plot(kind='bar', color="skyblue", edgecolor="black")
plt.title("Nhiệt độ trung bình theo từng tháng (°C)")
plt.xlabel("Tháng")
plt.ylabel("Nhiệt độ trung bình (°C)")
plt.xticks(rotation=0)
plt.show()
```
