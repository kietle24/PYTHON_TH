```python
import numpy as np
temperatures = np.random.uniform(15, 40, 365)
avg_temp = np.mean(temperatures)
max_temp = np.max(temperatures)
min_temp = np.min(temperatures)
print(f"nhiet do trung binh: {avg_temp}")
print(f"nhiet do cao nhat: {max_temp}")
print(f"nhiet do thap nhat: {min_temp}")
```

```python
#3
import numpy as np

# Sinh dữ liệu nhiệt độ ngẫu nhiên cho 365 ngày
temperatures = np.random.uniform(15, 40, 365)

# Số ngày trong từng tháng (không xét năm nhuận)
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Chia dữ liệu theo tháng
monthly_means = []
start = 0
for days in days_in_month:
    end = start + days
    month_data = temperatures[start:end]
    monthly_means.append(np.mean(month_data))
    start = end

# Phân loại nhiệt độ theo mức
categories = []
for mean_temp in monthly_means:
    if mean_temp < 20:
        categories.append("Thấp")
    elif mean_temp <= 30:
        categories.append("Trung bình")
    else:
        categories.append("Cao")

# In kết quả
for i, (mean, cat) in enumerate(zip(monthly_means, categories), start=1):
    print(f"Tháng {i}: {mean:.2f}°C → {cat}")
```

```python
#4ve bieu do duong 
months = range(1,13) 
plt.plot(months, monthly_means, marker="o", color="b", linestyle="-", linewidth=2)
plt.xlabel("thang ")
plt.ylabel("nhiet do trung binh ") 
plt.title("bang thong ke nhiet do theo thang ", fontsize=6)
plt.xticks(months)
plt.grid(True, linestyle="-", alpha = 1)
for i, value in enumerate(monthly_means ,start = 1 ) : 
    plt.text(i ,value + 0.3 , f"{value:.1f}", ha= "center", fontsize  = 9 ) 
plt.show()
```

```python
# stock_analysis_to_pdf.py
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.figure import Figure

# -------------------------------
# 1. Đọc dữ liệu CSV
# -------------------------------
df = pd.read_csv("stock_data.csv", parse_dates=['Date'])

# Tính tỷ lệ thay đổi hàng ngày (%)
df['Daily_Change'] = df['Close'].pct_change() * 100

# Đảm bảo cột Date là kiểu datetime trước khi dùng .dt
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M')

# Tính thống kê trung bình và độ lệch chuẩn theo tháng
monthly_stats = df.groupby('Month')['Close'].agg(['mean', 'std']).reset_index()

# Ngày có tỷ lệ thay đổi lớn nhất
max_change_day = df.loc[df['Daily_Change'].abs().idxmax()]

# -------------------------------
# 2. Chọn dữ liệu của một tháng để vẽ biểu đồ
# (Ví dụ: tháng 09/2025)
# -------------------------------
month_data = df[df['Month'] == '2025-09']

# -------------------------------
# 3. Tạo PDF và lưu kết quả
# -------------------------------
with PdfPages('stock_analysis.pdf') as pdf:

    # ---- 3.1 Biểu đồ giá cổ phiếu ----
    plt.figure(figsize=(10,6))
    plt.plot(month_data['Date'], month_data['Close'], marker='o', linestyle='-', color='blue')
    plt.title('Biểu đồ giá cổ phiếu tháng 09/2025')
    plt.xlabel('Ngày')
    plt.ylabel('Giá đóng cửa')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    pdf.savefig()
    plt.close()

    # ---- 3.2 Kết quả phân tích ----
    fig = Figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    ax.axis('off')  # Ẩn trục
    text = f"""
KẾT QUẢ PHÂN TÍCH CỔ PHIẾU

1. Ngày có tỷ lệ thay đổi lớn nhất:
   {max_change_day['Date'].strftime('%Y-%m-%d')} - {max_change_day['Daily_Change']:.2f}%

2. Thống kê giá đóng cửa theo tháng:
{monthly_stats.to_string(index=False)}
"""
    ax.text(0, 1, text, fontsize=12, verticalalignment='top')
    pdf.savefig(fig)
    plt.close(fig)

print("Đã tạo file stock_analysis.pdf thành công.")
```

```python
import pandas as pd
import numpy as np

# Tạo dữ liệu tiêu thụ điện năng theo giờ (giả lập)
hours = list(range(24))
# Giả lập tiêu thụ điện năng theo kWh, ví dụ: thấp vào đêm, cao vào ngày
consumption = [np.random.uniform(0.5, 1.5) if h<6 else
               np.random.uniform(1.5, 3.0) if 6<=h<=17 else
               np.random.uniform(1.0, 2.5) for h in hours]

# Tạo DataFrame
df = pd.DataFrame({
    'Hour': hours,
    'Consumption': consumption
})

# Lưu file CSV
df.to_csv('electricity_data.csv', index=False)
print("Đã tạo file electricity_data.csv")

# Đọc dữ liệu từ CSV
df = pd.read_csv('electricity_data.csv')
print(df.head())
```

```python
# Tổng điện năng tiêu thụ
total_consumption = df['Consumption'].sum()
print(f"Tổng điện năng tiêu thụ trong ngày: {total_consumption:.2f} kWh")

# Giờ cao điểm
peak_hour = df.loc[df['Consumption'].idxmax()]
print(f"Giờ cao điểm: {int(peak_hour['Hour'])}h - {peak_hour['Consumption']:.2f} kWh")
```

```python
# Thêm cột khung giờ
def time_slot(hour):
    if 6 <= hour <= 11:
        return 'Sáng'
    elif 12 <= hour <= 17:
        return 'Trưa'
    elif 18 <= hour <= 23:
        return 'Tối'
    else:
        return 'Đêm'

df['Slot'] = df['Hour'].apply(time_slot)

# Tổng tiêu thụ theo khung giờ
slot_consumption = df.groupby('Slot')['Consumption'].sum().reset_index()
print("Tổng tiêu thụ theo khung giờ:")
print(slot_consumption)

# Khung giờ tiêu thụ nhiều nhất
max_slot = slot_consumption.loc[slot_consumption['Consumption'].idxmax()]
print(f"Khung giờ tiêu thụ nhiều nhất: {max_slot['Slot']} - {max_slot['Consumption']:.2f} kWh")
```

```python
%pip install mpld3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpld3

# -------------------------------
# 1. Tạo dữ liệu tiêu thụ điện năng theo giờ
# -------------------------------
hours = list(range(24))
consumption = [np.random.uniform(0.5, 1.5) if h<6 else
               np.random.uniform(1.5, 3.0) if 6<=h<=17 else
               np.random.uniform(1.0, 2.5) for h in hours]

df = pd.DataFrame({'Hour': hours, 'Consumption': consumption})

# -------------------------------
# 2. Tính tổng tiêu thụ và giờ cao điểm
# -------------------------------
total_consumption = df['Consumption'].sum()
peak_hour = df.loc[df['Consumption'].idxmax()]

# -------------------------------
# 3. Phân tích theo khung giờ
# -------------------------------
def time_slot(hour):
    if 6 <= hour <= 11:
        return 'Sáng'
    elif 12 <= hour <= 17:
        return 'Trưa'
    elif 18 <= hour <= 23:
        return 'Tối'
    else:
        return 'Đêm'

df['Slot'] = df['Hour'].apply(time_slot)
slot_consumption = df.groupby('Slot')['Consumption'].sum().reset_index()
max_slot = slot_consumption.loc[slot_consumption['Consumption'].idxmax()]

# -------------------------------
# 4. Vẽ biểu đồ cột
# -------------------------------
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(slot_consumption['Slot'], slot_consumption['Consumption'], color=['orange','green','blue','purple'])
ax.set_title('So sánh tiêu thụ điện giữa các khung giờ')
ax.set_xlabel('Khung giờ')
ax.set_ylabel('Tổng tiêu thụ (kWh)')
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Chuyển biểu đồ Matplotlib sang HTML tương tác
chart_html = mpld3.fig_to_html(fig)
plt.close(fig)

# -------------------------------
# 5. Tạo HTML báo cáo
# -------------------------------
html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <meta charset="UTF-8">
    <title>Báo cáo tiêu thụ điện năng</title>
</head>
<body>
    <h1>Báo cáo phân tích tiêu thụ điện năng trong ngày</h1>

    <h2>1. Tổng tiêu thụ điện năng</h2>
    <p>Tổng tiêu thụ trong ngày: <b>{total_consumption:.2f} kWh</b></p>
    <p>Giờ cao điểm: <b>{int(peak_hour['Hour'])}h</b> - {peak_hour['Consumption']:.2f} kWh</p>

    <h2>2. Tiêu thụ theo khung giờ</h2>
    <table border="1" cellpadding="5" cellspacing="0">
        <tr>
            <th>Khung giờ</th>
            <th>Tổng tiêu thụ (kWh)</th>
        </tr>
"""

# Thêm dữ liệu bảng vào HTML
for _, row in slot_consumption.iterrows():
    html_content += f"<tr><td>{row['Slot']}</td><td>{row['Consumption']:.2f}</td></tr>\n"

html_content += f"""
    </table>
    <p>Khung giờ tiêu thụ nhiều nhất: <b>{max_slot['Slot']}</b> - {max_slot['Consumption']:.2f} kWh</p>

    <h2>3. Biểu đồ tiêu thụ điện</h2>
    {chart_html}

</body>
</html>
"""

# Lưu HTML
with open("electricity_report.html", "w", encoding="utf-8") as f:
    f.write(html_content)

print("Đã tạo file electricity_report.html thành công.")
```
